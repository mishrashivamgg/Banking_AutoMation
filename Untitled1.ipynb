```python
import sqlite3
import time
from tkinter import *
from tkinter import messagebox
from tkinter.ttk import Combobox

# Create database and table if not exists
try:
    conobj = sqlite3.connect(database="bank.sqlite")
    curobj = conobj.cursor()
    curobj.execute(
        "CREATE TABLE IF NOT EXISTS accounts (acn INTEGER PRIMARY KEY AUTOINCREMENT, name TEXT, pass TEXT, email TEXT, mob TEXT, bal FLOAT, type TEXT, opendate TEXT)"
    )
    conobj.close()
    print("hello")
except Exception as e:
    print(f'Something went wrong: {e}')

# Initialize the main window
win = Tk()
win.state('zoomed')
win.configure(bg='pink')
win.resizable(width=False, height=False)
title = Label(win, text="Banking Automation", font=('arial', 50, 'bold', 'underline'))
title.pack()
dt = time.strftime("%d %b, %Y")
date = Label(win, text=f"{dt}", font=('arial', 20, 'bold'), bg='pink', fg='blue')
date.place(relx=.85, rely=.1)

def main_screen():
    frm = Frame(win)
    frm.configure(bg='sky blue')
    frm.place(relx=0, rely=.15, relwidth=1, relheight=.85)
    
    def forgotpass():
        frm.destroy()
        forgotpass_screen()
    
    def newuser():
        frm.destroy()
        newuser_screen()
        
    def login():
        acn = e_acn.get()
        pwd = e_pass.get()
        if len(acn) == 0 or len(pwd) == 0:
            messagebox.showwarning('Validation', "Empty fields are not allowed")
            return
            
        frm.destroy()
        welcome_screen()

    def clear():
        e_acn.delete(0, "end")
        e_pass.delete(0, "end")
        e_acn.focus()
        
    lab_acn = Label(frm, text="ACN", font=('arial', 20, 'bold'), bg='powder blue')
    lab_acn.place(relx=.3, rely=.1)
    
    e_acn = Entry(frm, font=('arial', 20, 'bold'), bd=5)
    e_acn.place(relx=.4, rely=.1)
    e_acn.focus()
    
    lab_pass = Label(frm, text="Pass", font=('arial', 20, 'bold'), bg='powder blue')
    lab_pass.place(relx=.3, rely=.2)
    
    e_pass = Entry(frm, font=('arial', 20, 'bold'), bd=5, show='*')
    e_pass.place(relx=.4, rely=.2)
    
    btn_login = Button(frm, text="login", font=('arial', 20, 'bold'), bd=5, command=login)
    btn_login.place(relx=.41, rely=.3)
    
    btn_clear = Button(frm, text="clear", font=('arial', 20, 'bold'), bd=5, command=clear)
    btn_clear.place(relx=.51, rely=.3)
    
    btn_fp = Button(frm, width=15, text="Forgot Password", font=('arial', 20, 'bold'), bd=5, command=forgotpass)
    btn_fp.place(relx=.4, rely=.43)
    
    btn_new = Button(frm, width=15, command=newuser, text="Open New Account", font=('arial', 20, 'bold'), bd=5)
    btn_new.place(relx=.4, rely=.55)

def forgotpass_screen():
    frm = Frame(win)
    frm.configure(bg='powder blue')
    frm.place(relx=0, rely=.15, relwidth=1, relheight=.85)
    
    def back():
        frm.destroy()
        main_screen()
        
    def clear():
        e_acn.delete(0, "end")
        e_mob.delete(0, "end")
        e_email.delete(0, "end")
        
    btn_new = Button(frm, width=15, text="back", font=('arial', 20, 'bold'), bd=5, command=back)
    btn_new.place(relx=0, rely=0)
    
    lab_acn = Label(frm, text="ACN", font=('arial', 20, 'bold'), bg='powder blue')
    lab_acn.place(relx=.3, rely=.1)
    
    e_acn = Entry(frm, font=('arial', 20, 'bold'), bd=5)
    e_acn.place(relx=.4, rely=.1)
    
    lab_email = Label(frm, text="Email", font=('arial', 20, 'bold'), bg='powder blue')
    lab_email.place(relx=.3, rely=.2)
    
    e_email = Entry(frm, font=('arial', 20, 'bold'), bd=5)
    e_email.place(relx=.4, rely=.2)
    
    lab_mob = Label(frm, text="Mobile", font=('arial', 20, 'bold'), bg='powder blue')
    lab_mob.place(relx=.3, rely=.3)
    
    e_mob = Entry(frm, font=('arial', 20, 'bold'), bd=5)
    e_mob.place(relx=.4, rely=.3)
    
    btn_sub = Button(frm, width=7, text="Submit", font=('arial', 20, 'bold'), bd=5)
    btn_sub.place(relx=.40, rely=.4)
    
    btn_clear = Button(frm, width=7, text="clear", font=('arial', 20, 'bold'), bd=5, command=clear)
    btn_clear.place(relx=.54, rely=.4)
    
def newuser_screen():
    frm = Frame(win)
    frm.configure(bg='powder blue')
    frm.place(relx=0, rely=.15, relwidth=1, relheight=.85)
    
    def back():
        frm.destroy()
        main_screen()
        
    def clear():
        e_name.delete(0, "end")
        e_pass.delete(0, "end")
        e_mob.delete(0, "end")
        e_email.delete(0, "end")
        cb_gender.set('')
        
    def newuser_db():
        name = e_name.get()
        pwd = e_pass.get()
        email = e_email.get()
        mob = e_mob.get()
        gender = cb_gender.get()
        bal = 0.0
        opendate = time.strftime("%d %b %Y, %A")
        
        try:
            conobj = sqlite3.connect(database="bank.sqlite")
            curobj = conobj.cursor()
            curobj.execute("INSERT INTO accounts (name, pass, email, mob, type, bal, opendate) VALUES (?, ?, ?, ?, ?, ?, ?)",
                           (name, pwd, email, mob, gender, bal, opendate))
            conobj.commit()
            conobj.close()
            messagebox.showinfo("New User", "Account Created Successfully")
        except Exception as e:
            messagebox.showerror("Error", f"Something went wrong: {e}")
        
    btn_new = Button(frm, width=15, text="back", font=('arial', 20, 'bold'), bd=5, command=back)
    btn_new.place(relx=0, rely=0)
    
    lab_name = Label(frm, text="Name", font=('arial', 20, 'bold'), bg='powder blue')
    lab_name.place(relx=.3, rely=.1)
    
    e_name = Entry(frm, font=('arial', 20, 'bold'), bd=5)
    e_name.place(relx=.4, rely=.1)
    
    lab_pass = Label(frm, text="Pass", font=('arial', 20, 'bold'), bg='powder blue')
    lab_pass.place(relx=.3, rely=.2)
    
    e_pass = Entry(frm, font=('arial', 20, 'bold'), bd=5, show='*')
    e_pass.place(relx=.4, rely=.2)
    
    lab_email = Label(frm, text="Email", font=('arial', 20, 'bold'), bg='powder blue')
    lab_email.place(relx=.3, rely=.3)
    
    e_email = Entry(frm, font=('arial', 20, 'bold'), bd=5)
    e_email.place(relx=.4, rely=.3)
    
    lab_mob = Label(frm, text="Mobile", font=('arial', 20, 'bold'), bg='powder blue')
    lab_mob.place(relx=.3, rely=.4)
    
    e_mob = Entry(frm, font=('arial', 20, 'bold'), bd=5)
    e_mob.place(relx=.4, rely=.4)
    
    lab_gender = Label(frm, text="Gender", font=('arial', 20, 'bold'), bg='powder blue')
    lab_gender.place(relx=.3, rely=.5)
    
    cb_gender = Combobox(frm, values=['Male', 'Female'], font=('arial', 20, 'bold'))
    cb_gender.place(relx=.4, rely=.5)
    
    btn_sub = Button(frm, width=7, text="Submit", font=('arial', 20, 'bold'), bd=5, command=newuser_db)
    btn_sub.place(relx=.40, rely=.6)
    
    btn_clear = Button(frm, width=7, text="clear", font=('arial', 20, 'bold'), bd=5, command=clear)
    btn_clear.place(relx=.54, rely=.6)

def welcome_screen():
    frm = Frame(win)
    frm.configure(bg='sky blue')
    frm.place(relx=0, rely=.15, relwidth=1, relheight=.85)
    
    def logout():
        frm.destroy()
        main_screen()
        
    def details():
        ifrm = Frame(frm, highlightbackground='black', highlightthickness=2)
        ifrm.configure(bg='white')
        ifrm.place(relx=.2, rely=.1, relwidth=.7, relheight=.5)
        
        lab_wel = Label(ifrm, text="This is Details Screen", font=('arial', 20, 'bold'), bg='white', fg='blue')
        lab_wel.pack()
        
    def update():
        ifrm = Frame(frm, highlightbackground='black', highlightthickness=2)
        ifrm.configure(bg='white')
        ifrm.place(relx=.2, rely=.1, relwidth=.7, relheight=.5)
        
        lab_wel = Label(ifrm, text="This is Update Screen", font=('arial', 20, 'bold'), bg='white', fg='blue')
        lab_wel.pack()
        
    def deposit():
        ifrm = Frame(frm, highlightbackground='black', highlightthickness=2)
        ifrm.configure(bg='white')
        ifrm.place(relx=.2, rely=.1, relwidth=.7, relheight=.5)
        
        lab_wel = Label(ifrm, text="This is Deposit Screen", font=('arial', 20, 'bold'), bg='white', fg='blue')
        lab_wel.pack()
        
    def withdraw():
        ifrm = Frame(frm, highlightbackground='black', highlightthickness=2)
        ifrm.configure(bg='white')
        ifrm.place(relx=.2, rely=.1, relwidth=.7, relheight=.5)
        
        lab_wel = Label(ifrm, text="This is Withdraw Screen", font=('arial', 20, 'bold'), bg='white', fg='blue')
        lab_wel.pack()
    
    def transfer():
        ifrm = Frame(frm, highlightbackground='black', highlightthickness=2)
        ifrm.configure(bg='white')
        ifrm.place(relx=.2, rely=.1, relwidth=.7, relheight=.5)
        
        lab_wel = Label(ifrm, text="This is Transfer Screen", font=('arial', 20, 'bold'), bg='white', fg='blue')
        lab_wel.pack()
    
    lab_wel = Label(frm, text="Welcome", font=('arial', 20, 'bold'), bg='powder blue')
    lab_wel.place(relx=0, rely=0)
    
    btn_logout = Button(frm, width=12, text="logout", font=('arial', 20, 'bold'), bd=5, command=logout)
    btn_logout.place(relx=.8, rely=0)
    
    btn_details = Button(frm, command=details, width=12, text="Details", font=('arial', 20, 'bold'), bd=5)
    btn_details.place(relx=0, rely=.11)
    
    btn_upt = Button(frm, command=update, width=12, text="Update", font=('arial', 20, 'bold'), bd=5)
    btn_upt.place(relx=0, rely=.22)
    
    btn_dpt = Button(frm, command=deposit, width=12, text="Deposit", font=('arial', 20, 'bold'), bd=5)
    btn_dpt.place(relx=0, rely=.33)
    
    btn_wtd = Button(frm, command=withdraw, width=12, text="Withdraw", font=('arial', 20, 'bold'), bd=5)
    btn_wtd.place(relx=0, rely=.44)
    
    btn_trf = Button(frm, command=transfer, width=12, text="Transfer", font=('arial', 20, 'bold'), bd=5)
    btn_trf.place(relx=0, rely=.55)

main_screen()
win.mainloop()
```
