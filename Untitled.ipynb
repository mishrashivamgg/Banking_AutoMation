```python
try:
    conobj=sqlite3.connect(database="bank.sqlite")
    curobj=conobj.cursor()
    curobj.execute("create table accounts(acn integer primary key autoincrement,name text,pass text,email text,mob text,bal float,type text,opendate text)")
    conobj.close()
    print("hello")
except:
    print('Something went worng')


win=Tk()
win.state('zoomed')
win.configure(bg='pink')
win.resizable(width=False,height=False)
title=Label(win,text="Banking Automation",font=('arial',50,'bold','underline'))
title.pack()
dt=time.strftime("%d %b,%Y")
date=Label(win,text=f"{dt}",font=('arial',0,'bold'),bg='pink',fg='blue')
date.place(relx=.85,rely=.1)

def main_screen():
    frm=Frame(win)
    frm.configure(bg='sky blue')
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    
    def forgotpass():
        frm.destroy()
        forgotpass_screen()
    
    def newuser():
        frm.destroy()
        newuser_screen()
        
    def login():
        global gacn
        gacn=e_acn.get()
        pwd=e_pass.get()
        if len(gacn)==0 or len(pwd)==0:
            messagebox.showwarning('Validation',"Empty fields are not allowd")
            return
        else:
            acn=e_acn.get()
            pwd=e_pass.get()
            conobj=sqlite3.connect(database="bank.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select * from accounts where acn=? and pass=?",(acn,pwd))
            tup=curobj.fetchone()
            conobj.close()
            if tup==None:
                messagebox.showerror("Login","Invalid ACN/PASS")
                
            else:
                frm.destroy()
                welcome_screen()
    def clear():
        e_acn.delete(0,"end")
        e_pass.delete(0,"end")
        e_acn.focus()
        
    lab_acn=Label(frm,text="ACN",font=('arial',20,'bold'),bg='powder blue')
    lab_acn.place(relx=.3,rely=.1)
    
    e_acn=Entry(frm,font=('arial',20,'bold'),bd=5)
    e_acn.place(relx=.4,rely=.1)
    e_acn.focus()
    
    lab_pass=Label(frm,text="Pass",font=('arial',20,'bold'),bg='powder blue')
    lab_pass.place(relx=.3,rely=.2)
    
    e_pass=Entry(frm,font=('arial',20,'bold'),bd=5,show='*')
    e_pass.place(relx=.4,rely=.2)
    
    btn_login=Button(frm,text="login",font=('arial',20,'bold'),bd=5,command=login)
    btn_login.place(relx=.41,rely=.3)
    
    btn_clear=Button(frm,text="clear",font=('arial',20,'bold'),bd=5,command=clear)
    btn_clear.place(relx=.51,rely=.3)
    
    btn_fp=Button(frm,width=15,text="Forgot Password",font=('arial',20,'bold'),bd=5,command=forgotpass)
    btn_fp.place(relx=.4,rely=.43)
    
    btn_new=Button(frm,width=15,command=newuser,text="open new account",font=('arial',20,'bold'),bd=5)
    btn_new.place(relx=.4,rely=.55)
    
def forgotpass_screen():
    frm=Frame(win)
    frm.configure(bg='powder blue')
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    
    def back():
        frm.destroy()
        main_screen()
        
    def clear():
        e_acn.delete(0,"end")
        
        e_mob.delete(0,"end")
        e_email.delete(0,"end")
        
    def forgotpass_db():
        acn=e_acn.get()
        email=e_email.get()
        mob=e_mob.get()
        
        conobj=sqlite3.connect(database="bank.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select * from accounts where acn=? and email=? and mob=?",(acn,email,mob))
        tup=curobj.fetchone()
        if tup==None:
            messagebox.showerror("Forgot Pass","Record not found")
            
        else:
            messagebox.showerror("Forgot Pass","You Pass={tup[0]}")
            conobj.close()
            
        
    btn_new=Button(frm,width=15,text="back",font=('arial',20,'bold'),bd=5,command=back)
    btn_new.place(relx=0,rely=0)
    
    lab_acn=Label(frm,text="ACN",font=('arial',20,'bold'),bg='powder blue')
    lab_acn.place(relx=.3,rely=.1)
    
    e_acn=Entry(frm,font=('arial',20,'bold'),bd=5)
    e_acn.place(relx=.4,rely=.1)
    
    lab_email=Label(frm,text="Email",font=('arial',20,'bold'),bg='powder blue')
    lab_email.place(relx=.3,rely=.2)
    
    e_email=Entry(frm,font=('arial',20,'bold'),bd=5)
    e_email.place(relx=.4,rely=.2)
    
    lab_mob=Label(frm,text="Mobial",font=('arial',20,'bold'),bg='powder blue')
    lab_mob.place(relx=.3,rely=.3)
    
    e_mob=Entry(frm,font=('arial',20,'bold'),bd=5)
    e_mob.place(relx=.4,rely=.3)
    
    btn_sub=Button(frm,width=7,text="Submit",font=('arial',20,'bold'),bd=5,command=forgotpass_db)
    btn_sub.place(relx=.40,rely=.4)
    
    btn_clear=Button(frm,width=7,text="clear",font=('arial',20,'bold'),bd=5,command=clear)
    btn_clear.place(relx=.54,rely=.4)
    
    
def newuser_screen():
    frm=Frame(win)
    frm.configure(bg='powder blue')
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    
    def back():
        frm.destroy()
        main_screen()
        
    def clear():
        e_name.delete(0,"end")
        e_pass.delete(0,"end")
        e_mob.delete(0,"end")
        e_email.delete(0,"end")
        cb_gender.delete(0,"end")
        
        
    def newuser_db():
        name=e_name.get()
        pwd=e_pass.get()
        email=e_email.get()
        mob=e_mob.get()
        gender=cb_gender.get()
        bal=0
        opendate=time.strftime("%d %b %Y,%A")
        
        match=re.fullmatch("[6-9][0-9]{9}",mob)
        if match==None:
            messagebox.showwarning("validation","Invalid format of mob")
            return
        
        match=re.fullmatch("[a-z][A-Z]+@[a-zA-Z]+\.[a-zA-Z]+",email)
        if match==None:
            messagebox.showwarning("validation","Invalid format of email")
            return

            
      

        import sqlite3
        conobj=sqlite3.connect(database="bank.sqlite")
        curobj=conobj.cursor()
        curobj.execute("insert into accounts(name,pass,email,mob,opendate,bal) values(?,?,?,?,?,?)",(name,pwd,email,mob,opendate,bal))
        conobj.commit()
        conobj.close()
        
        conobj=sqlite3.connect(database="bank.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select max(acn) from accounts")
        tup=curobj.fetchone()
        conobj.close()
        
        messagebox.showinfo("New User",f"Account Creat with ACN NO={tup[0]}")
        e_name.delete(0,"end")
        e_pass.delete(0,"end")
        e_mob.delete(0,"end")
        e_email.delete(0,"end")
        
        
        
        
        
    btn_new=Button(frm,width=15,text="back",font=('arial',20,'bold'),bd=5,command=back)
    btn_new.place(relx=0,rely=0)
    
    lab_name=Label(frm,text="Name",font=('arial',20,'bold'),bg='powder blue')
    lab_name.place(relx=.3,rely=.1)
    
    e_name=Entry(frm,font=('arial',20,'bold'),bd=5)
    e_name.place(relx=.4,rely=.1)
    
    lab_pass=Label(frm,text="Pass",font=('arial',20,'bold'),bg='powder blue')
    lab_pass.place(relx=.3,rely=.2)
    
    e_pass=Entry(frm,font=('arial',20,'bold'),bd=5,show='*')
    e_pass.place(relx=.4,rely=.2)
    
    lab_email=Label(frm,text="Email",font=('arial',20,'bold'),bg='powder blue')
    lab_email.place(relx=.3,rely=.3)
    
    e_email=Entry(frm,font=('arial',20,'bold'),bd=5)
    e_email.place(relx=.4,rely=.3)
    
    lab_mob=Label(frm,text="Mobial",font=('arial',20,'bold'),bg='powder blue')
    lab_mob.place(relx=.3,rely=.4)
    
    e_mob=Entry(frm,font=('arial',20,'bold'),bd=5)
    e_mob.place(relx=.4,rely=.4)
    
    lab_gender=Label(frm,text="Gender",font=('arial',20,'bold'),bg='powder blue')
    lab_gender.place(relx=.3,rely=.5)
    
    cb_gender=Combobox(frm,values=['----select----','Male','Female'],font=('arial',20,'bold'))
    cb_gender.place(relx=.4,rely=.5)
    
    btn_sub=Button(frm,width=7,text="Submit",font=('arial',20,'bold'),bd=5,command=newuser_db)
    btn_sub.place(relx=.40,rely=.6)
    
    btn_clear=Button(frm,width=7,text="clear",font=('arial',20,'bold'),bd=5,command=clear)
    btn_clear.place(relx=.54,rely=.6)

def welcome_screen():
    frm=Frame(win)
    frm.configure(bg='sky blue')
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    
    def logout():
        frm.destroy()
        main_screen()
        
    def details():
        ifrm=Frame(frm,highlightbackground='black',highlightthickness=2)
        ifrm.configure(bg='white')
        ifrm.place(relx=.2,rely=.1,relwidth=.7,relheight=.5)
        
        lab_wel=Label(ifrm,text="This is Details Screen",font=('arial',20,'bold'),bg='white',fg='blue')
        lab_wel.pack()
        
        conobj=sqlite3.connect(database="bank.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select opendate,bal,email,mob from accounts where acn=?",(gacn,))
        tup=curobj.fetchone()
        conobj.close()
        
        lab_opendate=Label(ifrm,text=f"Open Date:{tup[0]}",font=('arial',15,'bold'),bg='white')
        lab_opendate.place(relx=.2,rely=.12)
                           
        lab_bal=Label(ifrm,text=f"Balance:{tup[1]}",font=('arial',15,'bold'),bg='white')
        lab_bal.place(relx=.2,rely=.28)
                    
                         
        lab_email=Label(ifrm,text=f"Email:{tup[2]}",font=('arial',15,'bold'),bg='white')
        lab_email.place(relx=.2,rely=.48)
        
        lab_mob=Label(ifrm,text=f"Mob:{tup[3]}",font=('arial',15,'bold'),bg='white')
        lab_mob.place(relx=.2,rely=.68)
        
        
        
    def update():
        ifrm=Frame(frm,highlightbackground='black',highlightthickness=2)
        ifrm.configure(bg='white')
        ifrm.place(relx=.2,rely=.1,relwidth=.7,relheight=.5)
        
        conobj=sqlite3.connect(database="bank.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select name,pass,email,mob from accounts where acn=?",(gacn,))
        tup=curobj.fetchone()
        conobj.close()
        
        lab_wel=Label(ifrm,text="This is Update Screen",font=('arial',20,'bold'),bg='white',fg='blue')
        lab_wel.pack()
        
        lab_name=Label(ifrm,text="Name",font=('arial',20,'bold'),bg='white')
        lab_name.place(relx=.1,rely=.1)
    
        e_name=Entry(ifrm,font=('arial',20,'bold'),bd=5)
        e_name.place(relx=.1,rely=.2)
        e_name.insert(0,tup[0])
        
        lab_pass=Label(ifrm,text="Pass",font=('arial',20,'bold'),bg='white')
        lab_pass.place(relx=.1,rely=.4)
    
        e_pass=Entry(ifrm,font=('arial',20,'bold'),bd=5,show='*')
        e_pass.place(relx=.1,rely=.5)
        e_pass.insert(0,tup[1])
            
        lab_email=Label(ifrm,text="Email",font=('arial',20,'bold'),bg='white')
        lab_email.place(relx=.55,rely=.1)
    
        e_email=Entry(ifrm,font=('arial',20,'bold'),bd=5)
        e_email.place(relx=.55,rely=.2)
        e_email.insert(0,tup[2])
    
        lab_mob=Label(ifrm,text="Mobial",font=('arial',20,'bold'),bg='white')
        lab_mob.place(relx=.55,rely=.4)
    
        e_mob=Entry(ifrm,font=('arial',20,'bold'),bd=5)
        e_mob.place(relx=.55,rely=.5)
        e_mob.insert(0,tup[3])
        
        def update_db():
            name=e_name.get()
            pwd=e_pass.get()
            email=e_email.get()
            mob=e_mob.get()
            
            conobj=sqlite3.connect(database="bank.sqlite")
            curobj=conobj.cursor()
            curobj.execute("update accounts set name=?,pass=?,email=?,mob=? where acn=?",(name,pwd,email,mob,gacn))
            conobj.commit()
            conobj.close()
            
            messagebox.showinfo("Update","Record Update")
            welcome_screen()
            
        
        btn_update=Button(ifrm,width=12,text="Update",font=('arial',20,'bold'),bd=5,command=update_db)
        btn_update.place(relx=.65,rely=.7)
        
    def deposit():
        ifrm=Frame(frm,highlightbackground='black',highlightthickness=2)
        ifrm.configure(bg='white')
        ifrm.place(relx=.2,rely=.1,relwidth=.7,relheight=.5)
        
        lab_wel=Label(ifrm,text="This is Deposit Screen",font=('arial',20,'bold'),bg='white',fg='blue')
        lab_wel.pack()
        
        lab_amt=Label(ifrm,text="Amount",font=('arial',20,'bold'),bg='white')
        lab_amt.place(relx=.1,rely=.2)
    
        e_amt=Entry(ifrm,font=('arial',20,'bold'),bd=5)
        e_amt.place(relx=.3,rely=.2)
        
        def deposit_db():
            amt=float(e_amt.get())
            conobj=sqlite3.connect(database="bank.sqlite")
            curobj=conobj.cursor()
            curobj.execute("update accounts set bal=bal+? where acn=?",(amt,gacn))
            conobj.commit()
            conobj.close()
            
            messagebox.showinfo("Update",f"{amt}  Amount Deposited")
        
        btn_sub=Button(ifrm,width=12,text="Submit",font=('arial',20,'bold'),bd=5,command=deposit_db)
        btn_sub.place(relx=.3,rely=.4)
        
        
    def withdraw():
        ifrm=Frame(frm,highlightbackground='black',highlightthickness=2)
        ifrm.configure(bg='white')
        ifrm.place(relx=.2,rely=.1,relwidth=.7,relheight=.5)
        
        lab_wel=Label(ifrm,text="This is Withdraw Screen",font=('arial',20,'bold'),bg='white',fg='blue')
        lab_wel.pack()
        
        lab_amt=Label(ifrm,text="Amount",font=('arial',20,'bold'),bg='white')
        lab_amt.place(relx=.1,rely=.2)
    
        e_amt=Entry(ifrm,font=('arial',20,'bold'),bd=5)
        e_amt.place(relx=.3,rely=.2)
        
        def withdraw_db():
            amt=float(e_amt.get())
            
            conobj=sqlite3.connect(database="bank.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select bal from accounts where acn=?",(gacn,))
            tup=curobj.fetchone()
            avail_bal=tup[0]
            conobj.close()
            
            if avail_bal>amt:
                conobj=sqlite3.connect(database="bank.sqlite")
                curobj=conobj.cursor()
                curobj.execute("update accounts set bal=bal-? where acn=?",(amt,gacn))
                conobj.commit()
                conobj.close()
                messagebox.showinfo("Update",f"{amt}  Amount withdrawn")
            
            else:
                messagebox.showinfo("Withdraw","Insufficient bal")
                
        btn_sub=Button(ifrm,width=12,text="Submit",font=('arial',20,'bold'),bd=5,command=withdraw_db)
        btn_sub.place(relx=.3,rely=.4)
        
        
    
    def transfer():
        ifrm=Frame(frm,highlightbackground='black',highlightthickness=2)
        ifrm.configure(bg='white')
        ifrm.place(relx=.2,rely=.1,relwidth=.7,relheight=.5)
        
        lab_wel=Label(ifrm,text="This is Transfer Screen",font=('arial',20,'bold'),bg='white',fg='blue')
        lab_wel.pack()
        
        
        lab_amt=Label(ifrm,text="Amount",font=('arial',20,'bold'),bg='white')
        lab_amt.place(relx=.1,rely=.2)
    
        e_amt=Entry(ifrm,font=('arial',20,'bold'),bd=5)
        e_amt.place(relx=.3,rely=.2)
        
        
        lab_to=Label(ifrm,text="To",font=('arial',20,'bold'),bg='white')
        lab_to.place(relx=.1,rely=.4)
    
        e_to=Entry(ifrm,font=('arial',20,'bold'),bd=5)
        e_to.place(relx=.3,rely=.4)
        
        def transfer_db():
            to_acn=e_to.get()
            amt=float(e_amt.get())
            
            if to_acn==gacn:
                messagebox.showwarning("Transfer","To and From can't be same" )
                return
            
            conobj=sqlite3.connect(database="bank.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select bal from accounts where acn=?",(gacn,))
            tup=curobj.fetchone()
            avail_bal=tup[0]
            conobj.close()
            
            conobj=sqlite3.connect(database="bank.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select bal from accounts where acn=?",(to_acn,))
            tup=curobj.fetchone()
            conobj.close()
            
            if tup==None:
                messagebox.showwarning("Transfer","Invalid To ACN")
                return
            if avail_bal>=amt:
                conobj=sqlite3.connect("bank.sqlite")
                curobj=conobj.cursor()
                curobj.execute("update accounts set bal=bal+? where acn=?",(amt,to_acn))
                curobj.execute("update accounts set bal=bal+? where acn=?",(amt,gacn))
                conobj.commit()
                conobj.close()
                messagebox.showinfo("Transfer",f"{amt} transfer to ACN {to_acn}")
            
        btn_sub=Button(ifrm,width=12,text="Submit",font=('arial',20,'bold'),bd=5,command=transfer_db)
        btn_sub.place(relx=.3,rely=.54)
    
    conobj=sqlite3.connect(database="bank.sqlite")
    curobj=conobj.cursor()
    curobj.execute("select name from accounts where acn=?",(gacn,))
    tup=curobj.fetchone()
    conobj.close()
    
    lab_wel=Label(frm,text=f"Welcome,{tup[0]}",font=('arial',20,'bold'),bg='powder blue')
    lab_wel.place(relx=0,rely=0)
    
    btn_logout=Button(frm,width=12,text="logout",font=('arial',20,'bold'),bd=5,command=logout)
    btn_logout.place(relx=.8,rely=0)
    
    btn_details=Button(frm,command=details,width=12,text="Details",font=('arial',20,'bold'),bd=5)
    btn_details.place(relx=0,rely=.11)
    
    btn_upt=Button(frm,command=update,width=12,text="Update",font=('arial',20,'bold'),bd=5)
    btn_upt.place(relx=0,rely=.22)
    
    
    btn_dpt=Button(frm,command=deposit,width=12,text="Deposit",font=('arial',20,'bold'),bd=5)
    btn_dpt.place(relx=0,rely=.33)
    
    btn_wtd=Button(frm,command=withdraw,width=12,text="Withdraw",font=('arial',20,'bold'),bd=5)
    btn_wtd.place(relx=0,rely=.44)
    
    btn_trf=Button(frm,command=transfer,width=12,text="Transfer",font=('arial',20,'bold'),bd=5)
    btn_trf.place(relx=0,rely=.55)
    
    
    
main_screen()
win.mainloop()
```
